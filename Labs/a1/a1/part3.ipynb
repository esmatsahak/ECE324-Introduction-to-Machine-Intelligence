```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread("pink_lake.png")
plt.imshow(img)
plt.show()

img_add = img + [0.25, 0.25, 0.25]
plt.imsave("img_add.png", np.clip(img_add, 0, 1))

img_chan_0 = np.zeros(img.shape)
img_chan_1 = np.zeros(img.shape)
img_chan_2 = np.zeros(img.shape)
img_chan_0[0:img_chan_0.shape[0],0:img_chan_0.shape[1],0:1] = img[0:img.shape[0],0:img.shape[1],0:1]
img_chan_1[0:img_chan_1.shape[0],0:img_chan_1.shape[1],1:2] = img[0:img.shape[0],0:img.shape[1],1:2]
img_chan_2[0:img_chan_2.shape[0],0:img_chan_2.shape[1],2:3] = img[0:img.shape[0],0:img.shape[1],2:3]
plt.imsave("img_chan_0.png", img_chan_0)
plt.imsave("img_chan_1.png", img_chan_1)
plt.imsave("img_chan_2.png", img_chan_2)

img_gray = np.empty(img.shape)
img_gray[0:img_gray.shape[0],0:img_gray.shape[1],0:img_gray.shape[2]] = 0.299 * img[0:img.shape[0],0:img.shape[1],0:1] + 0.587 * img[0:img.shape[0],0:img.shape[1],1:2] + 0.114 * img[0:img.shape[0],0:img.shape[1],2:3] 
plt.imsave("img_gray.png", img_gray)

img_crop = img[0:int(img.shape[0]/2),0:img.shape[1],0:img.shape[2]]
plt.imsave("img_crop.png", img_crop)

img_flip_vert = np.empty(img.shape)
img_flip_vert[0:img_flip_vert.shape[0],0:img_flip_vert.shape[1],0:img_flip_vert.shape[2]] = img[img.shape[0]::-1,0:img.shape[1],0:img.shape[2]]
plt.imsave("img_flip_vert.png", img_flip_vert)
```
